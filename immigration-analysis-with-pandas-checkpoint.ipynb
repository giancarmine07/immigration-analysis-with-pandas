```python
immig = immig.drop(columns=['DATAFLOW','LAST UPDATE','freq','agedef','age','unit','sex',
                            'OBS_FLAG'])
immig.head(40)
```

#### <span style='color:Blue'>Replace country code with country name </span>

```python
immig = immig.replace(['BE','BG','CZ','DK','DE','EE','IE','EL','ES','FR','HR','IT','CY',
                       'LV','LT','LU','HU','MT','NL','AT','PL','PT','RO','SI','SK','FI',
                       'SE','IS','LI','NO','CH','UK','ME','MK','UA',],['Belgium',
                       'Bulgaria', 'Czechia','Denmark','Germany','Estonia','Ireland',
                        'Greece','Spain','France','Croatia','Italy','Cyprus','Latvia',
                       'Lithuania','Luxembourg','Hungary','Malta','Netherlands',
                       'Austria','Poland','Portugal','Romania','Slovenia','Slovakia',
                       'Finland','Sweden','Iceland','Liechtenstein','Norway','Switzerland',
                       'United Kingdom','Montenegro','North Macedonia','Ukraine'])
```

#### <span style='color:Blue'>Rename columns </span>

```python
immig = immig.rename(columns={"geo": "Country", "OBS_VALUE": "Value"})
```

```python
immig.head(20)
```

#### <span style='color:Blue'>Pivot of table useful for plotting</span>

```python
immig_pivot = immig.pivot(index='Country', columns='TIME_PERIOD', 
                          values='Value').fillna(0)
```

```python
immig_pivot
```

#### <span style='color:Blue'>Sum of immigration 2012-2021 by country</span>

```python
immig_by_country = immig_pivot.sum(axis=1).sort_values(ascending=False)
immig_by_country
```

#### <span style='color:Blue'>Bar plot immigration 2012-2021 by country</span>

```python
plot1 = immig_by_country.plot(kind="bar", title="Sum Immigration by Country 2012-2021 *Millions", figsize=(10,5))
```

#### <span style='color:Blue'>Bar plot sum of immigration(all countries) by year</span>

```python
plot_byyear = immig_pivot.sum(axis=0)
plot2 = plot_byyear.plot(kind="bar", title="Immigration in Europe by year 2012-2021 *Millions", figsize=(8,5))
angle_degrees = 60
start_point = (6.5, 2) 
arrow_length = 3
end_point = (start_point[0] + arrow_length * np.cos(np.radians(angle_degrees)),
             start_point[1] + arrow_length * np.sin(np.radians(angle_degrees)))

# Annotate with arrow
plot2.annotate('Covid', xy=end_point, xytext=start_point,
            arrowprops=dict(facecolor='red', arrowstyle='->'))
```
